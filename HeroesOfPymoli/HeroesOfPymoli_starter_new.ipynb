## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# *Establish the bins 
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]



# Add a colum named "Age Ranges", and then use pd.cut, based on"Age", to Bin the Purchasing Data
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
purchase_data
```

```python
# Run basic calculations for "age_purchase_total for each age group": No.1: Groupby "Age Ranges", and for each range, sum the total price
agegroup_purchase_total = purchase_data.groupby(["Age Group"]).sum()["Price"]
agegroup_purchase_total
```

```python
# Run basic calculation for "age_average" --> No.2: Groupby "Age Group", and for each Age Group, calculate the average using "Mean" function.
agegroup_purchase_average = purchase_data.groupby(["Age Group"]).mean()["Price"]
agegroup_purchase_average
```

```python
#Run basic calculation for "age Group_purchase_counts" --> No. 3: Group by "Age Group", and for each Age Group, calculate the total "COUNTS"
agegroup_purchase_counts = purchase_data.groupby(["Age Group"]).count()["Price"]
agegroup_purchase_counts
```

```python
#Calculate the "Average purchase per person within each Age Group"

# counts = users['Age_bin'].value_counts().sort_index() --> This is from the above cells, calculated for how many people (counts) are within each "Age Group".

avg_purchase_perperson_within_each_agegroup = agegroup_purchase_total / counts
# avg_purchase_perperson_within_each_agegroup = "$"+ str(avg_purchase_perperson_within_each_agegroup)
# avg_purchase_perperson_within_each_agegroup
```

```python
# Putting all the above information into a data frame

final_table = pd.DataFrame({'Purchase Count':agegroup_purchase_counts, 
              'Average Purchase Price':avg_purchase_perperson_within_each_agegroup, 
              'Total Purchase Value':agegroup_purchase_total,
              'Avg Total Purchase per Person':avg_purchase_perperson_within_each_agegroup})
final_table
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order

* Display a preview of the summary data frame



```python
retrieved_data = purchase_data[['Item ID','Item Name','Price']]
retrieved_data
```

```python
retrieved_data_2 = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
retrieved_data_2

# groupby = retrieved_data.groupby['Item ID','Item Name']
# groupby
```

```python
total_item_purchase = retrieved_data_2.groupby(["Item ID", "Item Name"]).sum()["Price"].map("${:,.2f}".format)

avg_item_purchase = retrieved_data_2.groupby(["Item ID", "Item Name"]).mean()["Price"].map("${:,.2f}".format)

item_purchase_counts = retrieved_data_2.groupby(["Item ID", "Item Name"]).count()["Price"]
```

```python
final_table_most_popular_items = pd.DataFrame({'Purchase Count':item_purchase_counts, 
              'Average Purchase Price':avg_item_purchase, 
              'Total Purchase Value':total_item_purchase})                                
```

```python
final_table_most_popular_items_sorted = final_table_most_popular_items.sort_values("Purchase Count", ascending=False)
final_table_most_popular_items_sorted
```
